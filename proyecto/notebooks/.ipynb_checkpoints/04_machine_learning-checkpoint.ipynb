```python
# Celda 1: configuraci√≥n e imports

%run ./00_template.py

import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

LAYER_COMUNAS = "comunas_rm_censo"

# Capa de comunas (para mapas)
comunas = gpd.read_file(RUTA_GPKG, layer=LAYER_COMUNAS)

# Rutas a los datos procesados
indicadores_path = PROCESSED_DATA / "indicadores_servicios.csv"
accesibilidad_path = PROCESSED_DATA / "accesibilidad_servicios.csv"

indicadores = pd.read_csv(indicadores_path)
accesibilidad = pd.read_csv(accesibilidad_path)

print("Notebook 04 ‚Äì Identificaci√≥n de desiertos de servicio")
print(f"Indicadores:   {indicadores_path}")
print(f"Accesibilidad: {accesibilidad_path}")
```

```python
# Celda 2: combinar oferta (Notebook 02) y accesibilidad (Notebook 03)

columnas_clave = ["cod_comuna", "comuna", "poblacion"]

base = indicadores.merge(
    accesibilidad,
    on=columnas_clave,
    how="inner",
    suffixes=("_oferta", "_acceso"),
)

print("Columnas disponibles en la tabla integrada:")
base.columns
```

```python
# Celda 3: definici√≥n de servicios y variables utilizadas

SERVICIOS_ANALISIS = {
    "salud": {
        "tasa_col": "tasa_establecimientos_salud_x10k",
        "dist_col": "dist_min_salud_km",
        "cov_col": "porc_cubierto_salud",
    },
    "supermercados": {
        "tasa_col": "tasa_osm_supermercados_x10k",
        "dist_col": "dist_min_supermercados_km",
        "cov_col": "porc_cubierto_supermercados",
    },
}

base[list({v for d in SERVICIOS_ANALISIS.values() for v in d.values()})].head()
```

```python
# Celda 4: c√°lculo de umbrales por servicio

umbrales = {}

for nombre, cfg in SERVICIOS_ANALISIS.items():
    tasa_col = cfg["tasa_col"]
    dist_col = cfg["dist_col"]
    cov_col = cfg["cov_col"]

    # Usamos solo filas con datos v√°lidos en las tres variables
    sub = base[[tasa_col, dist_col, cov_col]].dropna()

    umbrales[nombre] = {
        "tasa_p25": np.percentile(sub[tasa_col], 25),
        "dist_p75": np.percentile(sub[dist_col], 75),
        "cov_p25": np.percentile(sub[cov_col], 25),
    }

umbrales
```

```python
# Celda 5: creaci√≥n de banderas e √≠ndices de desierto por servicio

base_desiertos = base.copy()

for nombre, cfg in SERVICIOS_ANALISIS.items():
    tasa_col = cfg["tasa_col"]
    dist_col = cfg["dist_col"]
    cov_col = cfg["cov_col"]

    u = umbrales[nombre]

    col_baja_tasa = f"{nombre}_baja_tasa"
    col_alta_dist = f"{nombre}_alta_dist"
    col_baja_cov = f"{nombre}_baja_cobertura"
    col_score = f"score_desierto_{nombre}"
    col_flag = f"es_desierto_{nombre}"

    base_desiertos[col_baja_tasa] = (base_desiertos[tasa_col] <= u["tasa_p25"]).astype(int)
    base_desiertos[col_alta_dist] = (base_desiertos[dist_col] >= u["dist_p75"]).astype(int)
    base_desiertos[col_baja_cov] = (base_desiertos[cov_col] <= u["cov_p25"]).astype(int)

    base_desiertos[col_score] = (
        base_desiertos[[col_baja_tasa, col_alta_dist, col_baja_cov]].sum(axis=1)
    )

    base_desiertos[col_flag] = (base_desiertos[col_score] >= 2).astype(int)

cols_mostrar = [
    "cod_comuna",
    "comuna",
    "poblacion",
] + [
    c for c in base_desiertos.columns if c.startswith("score_desierto_") or c.startswith("es_desierto_")
]

base_desiertos[cols_mostrar].head()
```

```python
# Celda 6: √≠ndice agregado de desiertos de servicio

flags_desierto = [c for c in base_desiertos.columns if c.startswith("es_desierto_")]

base_desiertos["n_servicios_en_desierto"] = base_desiertos[flags_desierto].sum(axis=1)

# Ordenar de mayor a menor n√∫mero de desiertos
resumen_desiertos = base_desiertos[
    ["cod_comuna", "comuna", "poblacion", "n_servicios_en_desierto"] + flags_desierto
].sort_values("n_servicios_en_desierto", ascending=False)

resumen_desiertos.head(15)
```

```python
# Celda 7: mapa del n√∫mero de servicios en desierto por comuna

comunas_desiertos = comunas.merge(
    base_desiertos[["cod_comuna", "n_servicios_en_desierto"]],
    left_on="CUT_COM",
    right_on="cod_comuna",
    how="left",
)

fig, ax = plt.subplots(figsize=(9, 9))

comunas_desiertos.plot(
    column="n_servicios_en_desierto",
    ax=ax,
    legend=True,
    cmap="Reds",
    edgecolor="black",
    linewidth=0.3,
)

ax.set_title("N√∫mero de servicios en condici√≥n de desierto por comuna", fontsize=12)
ax.set_axis_off()

plt.tight_layout()
save_figure(fig, "mapa_indice_desiertos")
plt.show()
```

```python
# Celda 8: mapa espec√≠fico de desiertos de salud

comunas_salud = comunas.merge(
    base_desiertos[["cod_comuna", "es_desierto_salud"]],
    left_on="CUT_COM",
    right_on="cod_comuna",
    how="left",
)

fig, ax = plt.subplots(figsize=(9, 9))

comunas_salud.plot(
    column="es_desierto_salud",
    ax=ax,
    legend=True,
    cmap="OrRd",
    edgecolor="black",
    linewidth=0.3,
)

ax.set_title("Comunas en desierto de salud (1 = desierto, 0 = no desierto)", fontsize=12)
ax.set_axis_off()

plt.tight_layout()
save_figure(fig, "mapa_desierto_salud_binario")
plt.show()
```

```python
# Guardar tabla integrada con desiertos de servicio
desiertos_path = PROCESSED_DATA / "desiertos_servicios.csv"
base_desiertos.to_csv(desiertos_path, index=False)
print(f"Tabla de desiertos de servicio guardada en: {desiertos_path}")
```
